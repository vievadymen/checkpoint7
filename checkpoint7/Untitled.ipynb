```python
import pandas as pd
import graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
dataset=pd.read_csv("data/titanic-passengers.csv", sep=";")
df= dataset.copy()
df
```

```python
#Preprocessing our data
def preprocess_data(new_data):
    df['Age'].fillna(df['Age'].mean(),inplace=True)
    df.replace({'Sex':{'male': 1,'female':0}},inplace=True)
    df['Cabin']=df.Cabin.fillna('G6')
    df.replace({'Survived':{'Yes': 1,'No':0}},inplace=True)
    return new_data
data=preprocess_data(df)
data
```

```python
#features extraction
x=data.drop(["Survived", "Name", "Cabin", "Ticket", "Embarked"], axis=1)
y= data["Survived"]
```

```python
#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
```

```python
#applying tree algorithm
dtree = tree.DecisionTreeClassifier()  
dtree.fit(x_train, y_train)   #fitting our model
y_pred=dtree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
dot_data = tree.export_graphviz(dtree, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("data")
graph
```
